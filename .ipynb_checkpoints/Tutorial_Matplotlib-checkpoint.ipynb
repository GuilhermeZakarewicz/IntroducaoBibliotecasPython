Para melhor compreender os dados que estamos lidando e fazer inferências, é fundamental visualizar o dataset graficamente. A partir de visualizações dos dados, podemos investigar e entender melhor nossos objetivos, além de formular novas perguntas e repostas. 

A biblioteca [Matplotlib](www.matplotlib.org) é o módulo para gráficos mais popular entre os usuários Python. É possível plotar diferentes tipos de gráficos (lineares, distribuições, histogramas, tipo pizza, etc.) com alta qualidade e diferentes formatos de saída. 

Vamos começar explorando os básicos. 

````python
import numpy as np
import matplotlib.pyplot as plt
````

A seguir, iremos criar um gráfico simples de uma função linear.

````python
a = np.arange(2,6)
b = np.arange(7,11)

plt.plot(a,b)
plt.show()
````

Podemos definir um título para nosso gráfico, assim como rotular os eixos x e y.

`plt.title("titulo aqui")`

`plt.xlabel("rotulo eixo x")`

`plt.ylabel("rotulo eixo y")`

````python
x = np.linspace(0,2*np.pi,100)
y = np.sin(x)

plt.plot(x,y)
plt.title("Gráfico Seno")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.show()
````

É possível plotar mais de um gráfico simultaneamente. Para organizar nossa visualização, podemos usar a opção `label=`. E não podemos esquecer de usar a função `plt.legend()`

````python
x = np.linspace(0,2*np.pi,100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x,y1,label="seno")
plt.plot(x,y2,label="cosseno")
plt.title("Gráfico Seno e Cosseno")
plt.xlabel("Eixo x")
plt.ylabel("Eixo y")
plt.legend()
plt.show()
````

Podemos modificar os estilos do gráfico, o que dá mais liberdade para apresentar nossos dados.

````python
x = np.arange(0, 10, 0.5)
y1 = x
y2 = x**2
y3 = x**3

plt.plot(x,y1,"r",label="primeiro grau")
plt.plot(x,y2,"bs",label="segundo grau")
plt.plot(x,y3,"mo",label="terceiro grau")
plt.legend()
plt.show()
````

Para plotar dois gráficos diferentes na mesma figura, podemos usar a opção `subplot`.

```
Sintaxe: plt.subplots(nrows, ncols, index)
```

O argumento de três dígitos especifica o número de linhas/colunas e o índice do subplot.

````python
plt.subplot(1,2,1) # teremos uma figura com uma linha e duas colunas. Estamos plotando no primeiro índice
plt.plot(x,y1)
plt.title("Primeiro plot")

plt.subplot(1,2,2) # agora, estamos plotando no segundo índice
plt.plot(x,y2)
plt.title("Segundo plot")

plt.show()
````

Podemos mudar o tamanho da figura com a opção `plt.figure(figsize=(a,b))`.

Além do plot convencional, também podemos visualizar histogramas (`plt.bar`), distribuições (`plt.scatter`)...

````python
disciplina = ['sismica 1', 'sismica 2', 'analise de sinais', 'calculo 1', 'calculo 2']
n_alunos = [15, 12, 24, 72, 51]
````

````python
plt.figure(figsize=(16, 8))

plt.subplot(221)
plt.plot(disciplina, n_alunos)

plt.subplot(222)
plt.bar(disciplina, n_alunos)

plt.subplot(223)
plt.scatter(disciplina, n_alunos)

plt.subplot(224)
plt.stem(disciplina, n_alunos)

plt.show()
````

Para compreender a distribuição de uma variável numérica contínua, usamos o `plt.hist()`. 

Neste exemplo, também usaremos outras opções:

`plt.text(x_coord, y_coord, "texto")`: adiciona um texto em uma coordenada específica

`plt.xlim(inf,sup)`: limita o eixo x

`plt.ylim(inf,sup)`: limita o eixo y

`plt.grid(True)`: define um grid em nosso gráfico
 

````python
np.random.seed(17)

mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

n, bins, patches = plt.hist(x, 50, density=True, facecolor='g', alpha=0.75)

plt.xlabel('QI')
plt.ylabel('Probabilidade')
plt.title('Histograma de QI')
plt.text(60, .025, r'$\mu=100,\ \sigma=15$')
plt.xlim(40, 160)
plt.ylim(0, 0.03)
plt.grid(True)
plt.show()
````

Exemplo de um gráfico pizza usando `plt.pie()`:

````python
linguagem = ["Python", "C", "C++0", "R", "Java"]
usuarios = [276, 250, 249, 120, 49]

plt.figure(figsize=(6,6))
plt.pie(usuarios, labels=linguagem,autopct="%1.1f%%")
plt.show()
````

A biblioteca também permite a visualização de matrizes 2D por meio do método `plt.imshow()`. É possível mudar a paleta de cores com a opção `colorbar=`. [Confira uma lista das paletas de cores aqui](https://matplotlib.org/stable/tutorials/colors/colormaps.html).

````python
M = np.ones([20,40])

for i in range(M.shape[0]):
    for j in range(M.shape[1]):
        M[i,j] = i+j
        
for i in range(10):
    M[i+5,20] = 50
    
plt.imshow(M.T,aspect="auto")
plt.show()
````

A opção `plt.colorbar()` nos mostra a paleta de cores utilizada.

````python
plt.figure(figsize=(10,10))

plt.subplot(221)
plt.imshow(M.T, aspect="auto",cmap="gray")
plt.title("gray")
plt.colorbar()

plt.subplot(222)
plt.imshow(M.T, aspect="auto",cmap="seismic")
plt.title("seismic")
plt.colorbar()

plt.subplot(223)
plt.imshow(M.T, aspect="auto",cmap="copper")
plt.title("copper")
plt.colorbar()

plt.subplot(224)
plt.imshow(M.T, aspect="auto",cmap="inferno")
plt.title("inferno")
plt.colorbar()

plt.show()
````

O matplotlib também cria gráficos em três dimensões com o módulo `mpl_toolkits.mplot3d`.

````python
from mpl_toolkits.mplot3d import axes3d

x = np.random.randint(low=50,high=200,size=300) 
y = np.random.randint(low=0,high=50,size=300)
z = np.random.randint(low=200,high=350,size=300)
t = np.random.randint(low=0,high=500,size=300)

fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection="3d")
ax.scatter(x,y,z,c=t, cmap="hot")
plt.show()
````

````python
fig = plt.figure(figsize=(8,8))
ax = plt.axes(projection='3d')

zline = np.linspace(0, 15, 1000)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')

zdata = 15 * np.random.random(500)
xdata = np.sin(zdata) + 0.1 * np.random.randn(500)
ydata = np.cos(zdata) + 0.1 * np.random.randn(500)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='winter')

plt.show()
````

````python
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(projection='3d')

X, Y, Z = axes3d.get_test_data(0.01)

ax.plot_wireframe(X, Y, Z, rstride=10, cstride=10, color="k")

plt.show()
````
